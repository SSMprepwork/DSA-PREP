```python
def bubble_sort(arr):
    n=len(arr)
    for i in range(n):                              #Iterate over list-loop1
        swapped=False
        for j in range(n-1):                        #Compare Adjacent elements-loop2
            if arr[j]>arr[j+1]:
                arr[j],arr[j+1]=arr[j+1],arr[j]
                swapped=True
        if not swapped:
            break
    return arr

lst=[5,8,3,1,2]
sorted_lst=bubble_sort(lst)
print(sorted_lst)
    
     
```

```python
def selection_sort(arr):
    n=len(arr)
    for i in range(n):
        min=i
        for j in range(i+1,n):              #Since first element is already sorted
            if arr[j]<arr[min]:
                min=j
        arr[i],arr[min]=arr[min],arr[i]
    return arr

lst=[3,7,9,1,4]
sorted_lst=selection_sort(lst)
print(sorted_lst)
```

```python
def insertion_sort(arr):
    n=len(arr)
    for i in range(1,n):
        key=arr[i]
        j=i-1                               #points to previous element
        while j>=0 and arr[j]>key:
            arr[j+1]=arr[j]
            j-=1
        arr[j+1]=key
    return arr

lst=[6,9,3,5,1,2]
sorted_lst=selection_sort(lst)
print(sorted_lst)
```
